# Group and Visualize Data

```python
WIND_INDCR_FLAG=df.groupby('REGION_CODE')['CITY'].count() 
WIND_INDCR_FLAG.head()
```

this groupby function counts how many times a city appears in a region code. In this instance these number can be used to find how many airports/airfields there are in Alabama.

```python
WIND_INDCR_FLAG.plot()  
```

This Graph shows which City by region code in Alabama has the most airports/airfilds. Form the visualization I can see that soem cities habe a much higher amount of airfilds than other.

```python
WIND_INDCR_FLAG.nlargest(10).plot.bar() 
```

This plot takes the same data as the one above but showis it in an easier to read format. From this we can see that AGL,AWS, and ASO are the Cities with the higherst number of Airfilds.
